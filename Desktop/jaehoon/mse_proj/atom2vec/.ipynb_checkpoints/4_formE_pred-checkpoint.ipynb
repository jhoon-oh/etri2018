# Load data

```python
import re
import pymatgen as mg
from pymatgen.ext.matproj import MPRester
```

```python
with MPRester("WnMieZK6NaAloLRT") as m:
    entries = m.query({"nelements": {"$eq": 4}}, ["pretty_formula", "formation_energy_per_atom"])
    all_compounds = [[e['pretty_formula'], e['formation_energy_per_atom']] for e in entries]
    num_all_comp = len(all_compounds)
print (num_all_comp)
```

```python
def hasNumbers(inputString):
    return bool(re.search(r'\d', inputString))

def hasParenthesis(inputString):
    return bool(re.search(r'\(', inputString))

def unpeel_par(inputString):
    before_parenthesis = inputString.split('(')[0]
    parenthesis_in = re.findall('[A-Z][^A-Z]*', inputString.split('(')[1].split(')')[0])
    parenthesis_out = int(inputString.split('(')[1].split(')')[1])
    
    unpeel_str = ''
    if hasNumbers(parenthesis_in[0]):
        num_idx = re.search('[0-9]', parenthesis_in[0]).span()[0]
        num = int(parenthesis_in[0][num_idx:])
        unpeel_str += parenthesis_in[0][:num_idx] + str(num*parenthesis_out)
    else:
        unpeel_str += parenthesis_in[0] + str(parenthesis_out)
        
    if hasNumbers(parenthesis_in[1]):
        num_idx = re.search('[0-9]', parenthesis_in[1]).span()[0]
        num = int(parenthesis_in[1][num_idx:])
        unpeel_str += parenthesis_in[1][:num_idx] + str(num*parenthesis_out)     
    else:
        unpeel_str += parenthesis_in[1] + str(parenthesis_out)

    return before_parenthesis+unpeel_str
```

```python
dataset = []
for i in range(num_all_comp):
    if hasParenthesis(all_compounds[i][0]):
        unpeel = unpeel_par(all_compounds[i][0])
        tmp = re.findall('[A-Z][^A-Z]*', unpeel)
    else:
        tmp = re.findall('[A-Z][^A-Z]*', all_compounds[i][0])
    
    atoms_tmp = []
    num_tmp = []
    
    for j in range(len(tmp)):
        if hasNumbers(tmp[j]):
            num_idx = re.search('[0-9]', tmp[j]).span()[0]
            atoms_tmp.append(tmp[j][:num_idx])
            num_tmp.append(int(tmp[j][num_idx:]))
        else:
            atoms_tmp.append(tmp[j])
            num_tmp.append(1)
            
    if sorted(num_tmp) == [1, 1, 2, 6]:
        print (tmp, atoms_tmp, num_tmp)
```
